```python
#x = {'Ex': 5, 'Gd': 4, 'TA':3, 'Fa':2, 'Po': 1}

for House_trainNew["ExterQual"] in House_trainNew:
    House_trainNew.loc[House_trainNew["ExterQual"] == 'Ex', "ExterQual"] = 5
    House_trainNew.loc[House_trainNew["ExterQual"] == 'Gd', "ExterQual"] = 4
    House_trainNew.loc[House_trainNew["ExterQual"] == 'TA', "ExterQual"] = 3
    House_trainNew.loc[House_trainNew["ExterQual"] == 'Fa', "ExterQual"] = 2
    House_trainNew.loc[House_trainNew["ExterQual"] == 'Po', "ExterQual"] = 1
#House_trainNew.loc[House_trainNew['ExterQual'] == 'Ex', 'ExterQual'] = 5
#House_trainNew.loc[House_trainNew['ExterQual'] == 'Gd', 'ExterQual'] =4
```

```python
#changing the ordinal values in ExterQual to numeric values
HEx = House_trainNew['ExterQual'][0]
for i in HEx:
    if i == 'Ex':
        return 5
    elif i == "Gd":
        return 4
    elif i == "TA":
        return 3
    elif i == "Fa":
        return 2
    elif i == "Po":
        return 1
    else:
        return 0
```
